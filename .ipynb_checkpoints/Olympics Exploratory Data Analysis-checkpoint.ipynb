```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading datasets

athletes = pd.read_csv('D:/Workspace/Coding/Machine Learning/Olympics Data Analysis Using Python/athlete_events.csv')
regions = pd.read_csv('D:/Workspace/Coding/Machine Learning/Olympics Data Analysis Using Python/noc_regions.csv')
```

```python
#Joining the dataframes

athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
#Making the column name consistent

athletes_df.rename(columns = {'region' : 'Region', 'notes' : 'Notes'}, inplace = True);
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
#Checking null values

nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
athletes_df.isnull().sum()
```

```python
#Filtering Team India's details

athletes_df.query('Team == "India"').head(5)
```

```python
#Filtering top 10 countries participating

top_10_count = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_count
```

```python
#Plotting for top 10 countries

plt.figure(figsize=(12,6))
plt.title('Overall Participation by Country')
sns.barplot(x = top_10_count.index, y = top_10_count, palette = 'Set1')
```

```python
#Age distribution of the athletes

plt.figure(figsize=(12,6))
plt.title('Age Distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'blue', edgecolor = 'white');
```

```python
#WINTER SPORTS

winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#SUMMER SPORTS

summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
#Gender Data

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
#Plotting gender data

plt.figure(figsize = (12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle = 150, shadow = True);
```

```python
#Total Medals

athletes_df.Medal.value_counts()
```

```python
#Total number of female athletes in each olympics.

female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')] [['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
#Total number of female athletes in each olympics.

female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')] [['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
#Plotting women participation visually

sns.set(style = 'darkgrid')
plt.figure(figsize = (20,10))
sns.countplot(x = 'Year', data = womenOlympics, palette = 'tab10')
plt.title("Women Participation")
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize = (20,10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
#Gold Medal Athletes

goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals.head()
```

```python
#Gold medal from nan values

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#Gold medal beyond 60

goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] > 60]
sporting_event
```

```python
#Visualising the data

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals Athelets over 60 years.')
```

```python
#Gold medals from each country

goldMedals.Region.value_counts().reset_index(name = 'Medal').head(5)
```

```python
#Visualising the data

totalGoldMedals = goldMedals.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index', y = 'Medal', data = totalGoldMedals, height = 5, kind = 'bar', palette = 'rocket')
g.despine(left = True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medal per Country")
```

```python
#Rio Olympics

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head(10)
```

```python
#Visualising the data

plt.figure(figsize=(12,6))
sns.barplot(x=team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel("Country wise medal in year 2016")
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'])]
```

```python
#Height vs Weight ScatterPlot

plt.figure(figsize=(12,10))
axis = sns.scatterplot(x = 'Height', y = 'Weight', data = not_null_medals, hue = 'Sex')
plt.title("Height vs Weight of Olympics Medalists")
```
